# Desafio 1: Pesquisar aprendizado supervisionado, não supervisionado e  por reforço.

- Aprendizado Supervisionado: Quando ensinamos a máquina a fazer alguma coisa, como por exemplo a qual o valor de algum item mostrando diversas caracteristicas e valores de itens diversos;

- Aprendizado não supervisionado: Modelos de Classificação e Clusterização, por exemplo quando passamos caracteristicas de pessoas e não se tem uma coluna de algum resultado final. Assim, o algoritmo irá buscar caracteristica em comum e criar grupos de pessoas semelhantes;

- Aprendizado por Reforço: Muito utilizado para uma sequência de decisões, por meio de tentativas e erros os modelos são avaliados e as características dos melhores pontuados são passadas para frente descartando assim aqueles que tiveram um baixo desempenho, é a principal ferramenta utilizada em modelos de Deep Learning.
