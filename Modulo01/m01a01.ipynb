## Melhorando a apresentação dos dados


Primeiramente serão removidos os números que aparecem antes das letras

```python
a = df['Unidade da Federação'].str.split(' ', 1, expand = True)
df.insert(value = a[0], column = 'Número do Estado', loc = 0)
df['Unidade da Federação'] = a[1]

df.head()
```

Em seguida, será feito um novo plot ordenando os valores de forma decrescente, para que os estados que com os maiores valores tenham destaques

```python
ax = df.sort_values(by='2020/Ago',ascending=False).plot(
    x = 'Unidade da Federação',
    y = '2008/Ago',
    kind = 'bar',
    figsize = (16,9),
    cmap = 'cividis',
    
)

ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("R$ {x:,.2f}"))
ax.set_xlabel('')

plt.title('Valor por Estado - Agosto/2020', fontsize = 18)
plt.xticks(rotation = 70)
plt.legend('')
plt.show()
```

## Análise para o mês mais recente

Primeiramente devemos buscar uma forma de automatizar a identificação do ultimo mês do conjunto de dados, que provavelmente será a última columa dos dados.

```python
df.keys()
```

Podemos observar que a última coluna deste conjunto de dados é a coluna Total. Com isso, temos duas opções: buscar pela penúltima coluna ou então remover a coluna Total.

Porém, mais pra frente a Coluna Total pode ser útil, então será feita a busca pela penúltima coluna do dataframe

```python
df.columns[-2]
```

```python
last_month = df.columns[-2]
ax = df.sort_values(by=last_month,ascending=False).plot(
    x = 'Unidade da Federação',
    y = last_month,
    kind = 'bar',
    figsize = (16,9),
    cmap = 'cividis',
    
)

ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("R$ {x:,.2f}"))
ax.set_xlabel('')

plt.title('Valor por Estado - '+last_month, fontsize = 18)
plt.xticks(rotation = 70)
plt.legend('')
plt.show()
```
